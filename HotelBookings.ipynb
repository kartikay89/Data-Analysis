```
filtered_rating_DF = bookingData[~bookingData.rate.str.contains("dubai")]
filtered_rating_DFV1 = filtered_rating_DF[~filtered_rating_DF.rate.str.contains("Review score")]
```

**Bar plot to see the type of ratings given by users**

```
#Rating bar chart

plt.subplots(figsize=(19,6))
sns.countplot('rate', data=filtered_rating_DFV1[1:] ,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain', axis='y')
plt.title('Ratings given by users')
plt.show()
```

**Looking at the city where hotels are most booked.**

```
#Cities bar chart

plt.subplots(figsize=(30,6))
sns.countplot('city', data=filtered_rating_DFV1[1:] ,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain', axis='y')
plt.title('Cities')
plt.show()
```
